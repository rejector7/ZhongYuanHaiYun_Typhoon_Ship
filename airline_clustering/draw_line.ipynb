```python
import itertools
# for ports, paths in path_port_dict.items():
#     paths = np.array(paths)
#     for path in paths:
#         path = np.array(path)
#         lon = path[:,1]
#         lat = path[:,2]
#         my_map.plot(
#             lon,
#             lat,
#             # markerfacecolor = "red",
#             # marker = "-",
#         )
# shanghai -> haikou
# path = get_path_by_ports((54, 202))
# shanghai -> xiamen
china_port_list1 = list(range(43, 50))
china_port_list2 = list(range(149, 345))
china_port_list3 = list(range(1243, 1308))
china_port_list4 = list(range(2268, 2301))
china_ports = china_port_list1 + china_port_list2 + china_port_list3 + china_port_list4
# print(type(china_ports))
list1_permu = itertools.permutations(china_port_list1, 2)
available_china_ports_pair = 0
for ports in list1_permu:
    path = get_path_by_ports(ports)
    if path is None:
        continue
    # print(len(path))
    available_china_ports_pair += 1
    lon = path[:,0]
    lat = path[:,1]
    my_map.plot(
        lon,
        lat,
        # markerfacecolor = "red",
        # marker = "-",
        color = "red"
    )
    print(ports)
    print(lon[0], lat[0], lon[-1], lat[-1])

print("available_china_ports_pair:", available_china_ports_pair)
plt.title("Port", fontsize = 20)

my_map.shadedrelief()
```
